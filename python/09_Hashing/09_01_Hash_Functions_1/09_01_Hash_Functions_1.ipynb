Find
،
Insert
و
Delete

$i$

$i$

<h1><font color=#FF7500 id="section2" size=6>
معرفی تابع درهمسازی
</font></h1><p></p>
تابع درهسازی، تابعی مانند
$h:D\rightarrow L$
است که اعضای
مجموعه
$D$
(که اندازه دلخواه
دارند) را به مجموعه‌
با اندازه ثابت
$L$
نگاشت می‌دهد.
معمولا این توابع به گونه‌ای هستند که اندازه 
$D$
بسیار بزرگتر از اندازه 
$L$
است که مزایا و معایب خود را دارد.
<br>
برای مثال می‌توانیم رشته‌های ساخته شده با حروف انگلیسی را با یک تابع درهمسازی به مجموعه اعداد یک بازه خاص بنگاریم به این صورت که به هر رشته، تعداد حروف آن رشته است نسبت می‌دهیم. البته در ادامه خواهیم دید که تابع درهمسازی‌ای مانند مثال بالا کاربرد چندانی ندارد و یک تابع درهمسازی خوب باید ویژگی‌های خاصی را دارا باشد. 
<br>
توابع درهمساز دامنه کاربردهای گسترده‌ای در مهندسی کامپیوتر دارند
از جمله:
<ul>
    <li> داده ساختارهایی چون جدول درهمسازی </li>
    <li> سیستم‌های رمزنگاری </li>
    <li> مبادلات اینترنتی </li>
    <li>سیستم‌های تعیین هویت دیجیتال مانند تشخیص اثر انگشت </li>
    <li> بررسی صحت فایل‌ها </li>
</ul>

<br>
حل مسائلی مشابه مساله قبلی هم از کاربردهای پایه‌ای درهمسازی است.
راه‌حل این سوال را در ادامه می‌آوریم:

<br>
<font color=#FF7500><b>
راه‌حل:</b>
</font>
فرض کنید در مساله قسمت قبل تابعی داشته باشیم که اعداد بین
$۰$
و
$۱۰^{۱۲}$
را به اعداد بین
$۰$
تا
$۱۰^{۴}$
نگاشت کند،با این تضمین که احتمال نگاشت شدن دو عدد متفاوت به یک عدد در این بازه قابل صرف نظر باشد(همچنین فرض کنید هزینه نگاشت کردن کم و قابل صرف نظر باشد که معمولا همینطور است).
حال با کمک این تابع، می‌توانیم ایده مساله اول را به مساله بزرگتر تعمیم دهیم به گونه‌ای که برای درج، یافتن و حذف یک عنصر مانند
$x$
کافیست به خانه
$h(x)$
از آرایه
$۱۰^{۴}$
عضوی رجوع کنیم.
<br>
قطعه کد زیر پیاده‌سازی این مسئله را آورده است:
<p></p>
</font></div>

```python
hash_size = 10**4 

def hash_function(x):
    return x % hash_size

class my_set:
    table = [0] * hash_size 
    
    def insert(self, x):
        self.table[hash_function(x)] = 1 
                
    def exists(self, x):
        return self.table[hash_function(x)] == 1
    
    def remove(self, x):
        self.table[hash_function(x)] = 0
        
s = my_set() 
s.insert(10)
s.insert(245)
print(s.exists(124))
s.insert(8653)
s.insert(12543252)
print(s.exists(10))
s.remove(10)
print(s.exists(10))
print(s.exists(12543252))
print(s.exists(12543251))
print(s.exists(3252))
```

$D$

$L$
مجموعه اندیسهای آرایه است.

<h2><font color=#FF7500 id="section3" size=5>
روش باقیمانده تقسیم
</font></h2>
در این روش به هر عدد باقی‌مانده تقسیمش بر
یک عدد ثابت مانند
$m$
را نسبت می‌دهیم.
این دقیقا همان کاری است که در مثال قبلی انجام دادیم
.
<br/>
$$ h(x) = x \:mod\: m $$
<p></p>
انتخاب مقدار
مناسبه برای
$m$
از اهمیت بالایی برخوردار است
.
به عنوان مثال اگر
$m$
توانی از دو باشد
خروجی تابع درهمساز برابر
$lg(m)$
بیت کم ارزش
$m$
می شود
و
در نتیجه بقیه بیت‌های عدد در خروجی تابع درهمساز نقشی ندارند.
به طور کلی بهتر است
$m$
یک عدد اول و دور از توان‌های دو باشد
. 
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<h2><font color=#FF7500 size=5>
روش ضرب
</font></h2>
یک استراتژی دیگر روش ضرب است که دو مرحله دارد. ابتدا داده‌ی ورودی
ضرب در یک عدد ثابت
$A$
که
$0 < A < 1$
شده و قسمت اعشاری آن گرفته می‌شود.
سپس این قسمت اعشاری در عدد صحیح
$m$
ضرب شده و قسمت صحیح آن گرفته می‌شود.
$$ h(x) = \lfloor m ( x A \:mod\: 1 ) \rfloor $$
که منظور از باقی‌مانده به ۱
همان قسمت اعشاری عدد بوده.
خوبی این روش این است که دیگر انتخاب عدد
$m$
اینقدر مهم نیست. معمولا در این روش
$m$
را توانی از ۲ گرفته و
$A$
را به صورت عدد گویای
$\frac{s}{2^w}$
می‌گیریم که
$s$
یک عدد صحیح در بازه‌ی
$0 < s < 2^w$
است.
</font></div>

<h2><font color=#FF7500 size=5>
روش همگانی - Universal hashing
</font></h2>
می‌توان برای
درهمسازی استفاده کرد، بالاخره امکان دارد اعداد ورودی به صورتی
انتخاب شوند که همه به یک خانه نگاشته شوند.

یک روش برای مقابله با این مشکل این است که
تعداد زیادی تابع درهمسازی آماده داشته باشیم
و در ابتدای اجرای برنامه یکی از این توابع را
به صورت مستقل از اعداد ورودی به صورت تصادفی
انتخاب کرده و در ادامه‌ی برنامه از این تابع استفاده کنیم.
به این کار روش همگانی درهم‌سازی گفته می‌شود.
<br/>
برای مثال اگر از روش ضرب استفاده می‌کنیم،
می‌توانیم تعداد زیادی زوج عدد
$m,A$
داشته باشیم و در هنگام اجرای برنامه یکی را انتخاب کرده
و باقی کارها را براساس آن انجام دهیم.
<br/>
ثابت می‌شود که این روش به طور متوسط حتی روی بدترین ورودی ممکن
روش کارآمدی است.
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<h1><font color=#FF7500 id="section4" size=6>
برخورد و روشهای مقابله با آن
</font></h1>
به وضوح نمی‌توانیم یک نگاشت یک به یک از یک مجموعه با کاردینالیتی
$C_1$
به
مجموعه‌ای با کاردینالیتی
$C_2$
داشته باشیم که
$C_1 > C_2$
.
پس
حتما وقتی از تابع درهمساز
$h$
برای نگاشت مجموعه
$D$
به
مجموعه
$L$
که
$|D| > |L|$
استفاده می کنیم، حتما عناصر متمایزی از
$D$
وجود دارند که یه یک عنصر واحد در
$L$
نگاشت شده‌اند
.
<br/>
به
عنوان مثال،
تابع درهمسازی ارائه شده در قسمت قبل، هر دو عدد
$۲۰۹۸۷$
و
$۱۰۹۸۷$
را به عدد
$۹۸۷$
می‌نگارد
(به قطعه کدی که در ادامه آمده توجه کنید).
حال اگر بخواهیم همزمان این دو عدد
را در مجموعه داشته باشیم به مشکل می‌خوریم.
به این اتفاق برخورد یا
collision
می گوییم.
حل کردن مشکل برخورد از مسائل پایه‌ای در طراحی توابع درهمسازی است.
در عمل، ما دوست داریم از تابع درهمسازی‌ای استفاده کنیم که تعداد برخوردهای آن در سناریوهای مدنظر ما کمینه باشد.

<h2><font color=#FF7500 size=5>
روشهای مقابله
</font></h2>
تا به حال سعی می‌کردیم تا توابعی پیدا کنیم
که احتمال برخورد را کم کنیم.
اما در هر صورت
اگر بخواهیم
تعداد زیادی عدد را در یک جدول با اندازه‌ی کوچک
ذخیره کنیم، حتما برخورد خواهیم داشت.
برای حل این مشکل روشهای زیادی وجود دارد از جمله:
		<ul>
			<li>زنجیره‌سازی</li>
			<li>آدرس‌دهی آزاد</li>
		</ul>
<br>
که در ادامه این دو روش را توضیح می‌دهیم.

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<h2><font color=#FF7500 size=5>
زنجیره‌سازی
</font></h2>
در این روش، هر خانه‌ی جدول را
یک لیست پیوندی‌ می‌گیریم.
حال هر عدد را که می‌خواهیم به خانه‌ای
اضافه کنیم، خود عدد را در انتهای لیست آن خانه
اضافه می‌کنیم. برای چک کردن وجود هم
باید کل لیست آن خانه را بگردیم
تا این که یا لیست تمام شود یا
عدد مورد نظر را پیدا کنیم.
<br/>
<div style="text-align:center">
    <img src="src/images/chain_making.png" style="width:50%">
    <font size=3> شکل۳: زنجیره‌سازی در جدول درهمسازی </font>
</div>
<br>
مشکل این روش داینامیک بودن سایز داده ساختار است.
<br/>
اگر طول یک لیست پیوندی خیلی زیاد شد می‌توانیم از هش و یا درخت دودویی جست‌و‌جو  به جای لیست پیوندی استفاده کنیم.
<br/>
این روش شاید به نظر کارا نیاید
اما دقت کنید که فرض کرده بودیم
قرار نیست تعداد زیادی برخورد داشته باشیم.
<br>
کد این جدول درهمسازی تعمیم‌یافته در ادامه آمده است:
</font></div>

```python
class my_chained_set:
    table = [[] for _ in range(hash_size)]

    def insert(self, x):
        if not self.exists(x):
            self.table[hash_function(x)].append(x) # O(1)

    def exists(self, x):
        return x in self.table[hash_function(x)] # O(size of list)

    def remove(self, x):
        self.table[hash_function(x)].remove(x)  # O(size of list)

s = my_chained_set()
s.insert(10)
s.insert(245)
print(s.exists(10))

s.insert(10987)
print(s.exists(987))  # right answer!
print(s.exists(10987))
```

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<h2><font color=#FF7500 size=5>
آدرس‌دهی باز
</font></h2>
در این روش، برعکس روش قبل، اعداد را در همان
خانه‌های جدول نگه‌داری می‌کنیم،
با این تفاوت که اگر به خانه‌ی اصلی رفتیم و پر بود،
به سراغ خانه‌ی (مشخص) دیگری می‌رویم.
مثلا فرض کنید که همیشه درست دنبال خانه‌ی بعدی برویم
و اگر آنجا خالی بود، عددمان را در آنجا قرار دهیم،
اگر نه باز هم ادامه دهیم تا به یک خانه‌ی خالی برسیم.
<br/>
اگر فرض کنیم که اندازه‌ی جدول خیلی بزرگ‌تر از تعداد عناصر داخل آن باشد،
به امکان زیاد خیلی نیاز نمی‌شود که
دنبال خانه‌ی دیگری برویم.
حال برای پیدا کردن یک عضو چه باید بکنیم؟
باید ابتدا سراغ خانه‌ی اصلی برویم و اگر خالی بود که
عدد در جدول نیست و اگر هم خود عدد در داخلش
قرار داشت که عدد پیدا شده و وجود دارد.
اگر خالی نبود ولی عدد ما هم در آن نبود،
باید به سراغ خانه‌ی بعدی برویم
تا وقتی که به یک خانه‌ی خالی برسیم.
برای حذف کردن هم ابتدا باید عنصر پیدا شود، سپس تا آخر دنباله‌ی
خانه‌های خالی رفته و آخرین عدد از این مجموعه (اگر وجود داشت)
به جای خودش قرار دهیم.
<br/>
البته در این روش
اگر خود خانه خالی بود، حتما به سراغ خانه‌ی
بعدی نمی‌رویم،
بلکه تابع درهمساز ما باید دو ورودی بگیرد
$h(x,i)$
و ابتدا ما به خانه‌ی
$h(x,0)$
می‌رویم. اگر پر بود به سراغ
$h(x,1)$
و
$h(x,2)$
و
...
می‌رویم تا یک خانه‌ی خالی پیدا کنیم.
<br/>
رسیدن به خانه‌ی خالی در جستجو برای یک کلید به معنای نبود این کلید در جدول می‌باشد.
<br>
<font color=#FF7500> 
<b>
تمرین ۱: 
</b>
</font>
آیا الگوریتم گفته شده برای حالتی که یک یا چند عنصر را از داده ساختار حذف کرده‌ایم هم جواب می‌دهد؟ اگر به مشکل برمیخورد راهی برای حل این مشکل ارائه دهید.
<br>
<br>
برای ساختن این تابع معمولا یکی از سه روش زیر را استفاده می‌کنیم:
<h3><font color=#FF7500 size=5>
۱- کاوش خطی
</font></h3>
این روش تقریبا همان روش رفتن به خانه‌ی بعدی است.
روشی که در آن ترتیب بررسی ثابت و معمولاً با قدم‌های به اندازه 1 می‌باشد. به این صورت که
اگر اعضای جدول اعداد صفر تا
$m$
باشند، تابع درهم‌ساز جدید را بر اساس تابع درهم‌ساز قبلی به صورت زیر می‌سازیم:
$$h(x,i) = (h'(x) + i) \:mod\: m$$
اما در این روش معمولا مشکلی به نام
دسته‌بندی اولیه (Primary Clustering)
پیش می‌آید که یعنی دنباله‌های پشت‌سر هم طولانی ساخته شده
و در نتیجه زمان جستجو افزایش می‌یابد.
<br/>
امکان وجود برخورد همچنین در این روش وجود دارد.
<h3><font color=#FF7500 size=5>
۲- کاوش مربعی
</font></h3>
روشی که در آن ترتیب بررسی تابعی از درجه دوم می‌باشد.
در این روش فرم تابع مورد استفاده به صورت زیر خواهد بود
$$h(x,i) = (h'(x) + c_1i + c_2i^2) \:mod\: m$$
اما باز هم اگر دو دسته تابع درهم‌ساز اولیه‌شان یکی باشد
باز هم دنباله‌‌شان یکسان خواهد بود.
به این مشکل نیز
دسته‌بندی ثانوی (Secondary Clustering)
می‌گویند.
<h3><font color=#FF7500 size=5>
۳- درهم‌سازی دوتایی
</font></h3>
حالتی که توالی جستجو برای هر کلید ثابت است اما به وسیله یک درهم‌ساز دیگر محاسبه می‌شود.
در این روش
برای حل مشکلات دو روش قبلی
از دو تابع درهم‌ساز متفاوت به صورت زیر استفاده می‌شود:
$$h(x,i) = (h_1(x) + ih_2(x)) \:mod\: m$$
<br/>
می‌توان گفت به طور تقریبی هیچ برخوردی در این روش اتفاق نمی‌افتد.
