<div id="queue">صف</div>
<p></p>
</font>
<hr>
صف داده ساختاری است که عنصر جدید فقط در انتهای آن اضافه می شود و حذف عنصر تنها از اول آن اتفاق می افتد. مانند یک صف نانوایی که هر کسی وارد می شود به انتهای صف می رود و هر کسی بعد از گرفتن نان از ابتدای صف خارج می شود.
<br>
<p></p>
صف باید از دستورات زیر پشتیبانی کند:
<ul style="margin-top:0">
<li>$enqueue(x)$: عنصر x را در انتهای صف درج می کند.</li>
<li>$dequeue()$:اولین عنصر را حذف می کند و آن را باز می گرداند.</li>
<li>$front()$: اولین عنصر صف  را باز می گرداند.</li>
<li>$size()$: تعداد عناصر موجود در صف  را باز می گرداند.</li>
<li>$is\_empty()$: خالی بودن یا نبودن صف را مشخص می کند.</li>
<li>$is\_full()$: پر بودن یا نبودن صف را مشخص می کند.</li>
</ul>
توجه:
تمامی  این اعمال در $O(1)$ 
انجام می‌شوند.

<p></p>
پیاده سازی صف معمولا به کمک آرایه یا لیست پیوندی که در ادامه با آن آشنا خواهید شد انجام میشود.
<p></p>


<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<font size=6>
<p></p>
پیاده سازی صف به کمک آرایه
</font>
<p></p>
برای پیاده سازی صف به کمک آرایه ابتدا یک آرایه 
و دو متغیر که یکی ابتدای صف 
($first$)
و دیگری تعداد عناصری که در صف قرار دارند
($num$)
را نشان می‌دهد، تعریف می کنیم .
<br>
هنگامی که عنصر جدیدی اضافه می شود آن را در انتهای صف 
قرار می دهیم و سپس مقدار 
$num$
را یکی افزایش می دهیم. برای حذف کردن عنصر ابتدایی صف نیز مقدار 
$num$ را یکی کم می کنیم
و همچنین مقدار 
$first$
را یکی افزایش می دهیم.
    (توجه: البته در توضیح بالا باید در نظر داشته باشیم که از آرایه به صورت دوری استفاده می کنیم برای مثال اگر سایز آرایه برابر 5 باشد، عنصر بعد از عنصر پنجم، عنصر اول است.)

```python
class Queue:
    # max_size: size of Q
    # Q: Array 
    def __init__(self, max_size):
        self.max_size = max_size
        self.Q = [0] * max_size
        self.num = 0
        self.first = 0
    
    def enqueue(self, item):
        if self.num >= self.max_size:
            raise Exception("Queue overflow")
        self.Q[(self.num + self.first) % self.max_size] = item
        self.num += 1
        
        
    def dequeue(self):
        if self.num == 0:
            raise Exception("Queue empty")
        item = self.Q[self.first]
        self.first = (self.first + 1) % self.max_size
        self.num -= 1
        return item
    
     
    def front(self):
        if self.num == 0:
            raise Exception("Queue empty")
        return self.Q[self.first]
    
    def is_empty(self):
        return self.num == 0
    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.max_size
```

```python
#Example

q=Queue(10) # (front of queue)[](back of queue)
q.enqueue("ra'na") # ["ra'na"]
q.enqueue("vez") # ["ra'na", "vez"]
q.enqueue("Arya") # ["ra'na", "vez", "Arya"]
print("queue size is: ",q.size())
print(q.dequeue(), "left the queue") # ["vez", "Arya"]
print("front of queue is:",q.front())
q.enqueue("milda") # ["vez", "Arya", "milda"]
q.dequeue() # ["Arya","milda"]
q.dequeue() # ["milda"]
q.dequeue() # []
print("It was a queue")
```

<div id="stack">پشته</div>
<p></p>
</font>
<hr>
در بعضی موارد عناصر تنها از یک طرف اضافه می شوند و از همان طرف نیز خارج می شوند.
به عنوان مثال ظرف های کثیفی را در نظر بگیرید 
که بر روی هم قرار دارند  و قصد شستن آن ها را دارید  در هر مرحله اگر بخواهید ظرفی را بردارید بالاترین آن ها را انتخاب می کنید و آن را می شویید و اگر ظرف کثیفی را بخواهید 
به آن ها اضافه کنید در بالای تمام آن ها قرار می دهید.
در واقع در این موارد عنصری که دیر تر از همه اضافه شده است زودتر از همه خارج میگردد.
به داده ساختاری که این گونه موارد را مدل می کند، پشته می گویند.
<p></p>
یک پشته باید از توابع زیر پشتیبانی کند:
<ul style="margin-top:0">
<li>$push(x)$: $x$ را به بالای پشته اضافه می کند.</li>
<li>$pop()$: عنصر بالای پشته را حذف می کند و آن را بازمی گرداند.</li>
<li>$top()$: عنصر بالای پشته را باز می گرداند.</li>
<li>$size()$: تعداد عناصر موجود در پشته را باز می گرداند.</li>
<li>$is\_empty()$: خالی بودن پشته را مشخص می کند.پ</li>
<li>$is\_full()$: پر بودن پشته را مشخص می کند.</li>
</ul>
تمام اعمال در $O(1)$ 
انجام می‌شوند.
<p></p>
پیاده سازی پشته نیز به کمک آرایه یا لیست پیوندی انجام میشود.
</font></div>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<font size=6>
<p></p>
پیاده سازی پشته به کمک آرایه
</font>
<p></p>
برای پیاده سازی پشته با آرایه از یک متغیر 
$num$
استفاده می کنیم که همیشه به عنصربالای پشته اشاره می کند.

```python
class Stack:
    def __init__(self, max_size):
        self.max_size = max_size        # Size of stack
        self.S = [0] * max_size     # Stack array
        self.num = 0            # Number of elements in Stack
    
    def push(self, item):
        if self.num >= self.max_size:
            raise Exception("Stack overflow")
        self.S[self.num] = item
        self.num += 1
        
    def pop(self):
        if self.num == 0:
            raise Exception("Stack empty")
        self.num -= 1
        return self.S[self.num]
    
    def top(self):
        if self.num == 0:
            raise Exception("Stack empty")
        return self.S[self.num-1]

    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.max_size
    
    def is_empty(self):
        return self.num == 0
```

```python
#Example

st=Stack(10) # (top of stack)[]
st.push(10) # [10]
st.push(30) # [30, 10]
st.push(1.5) # [1.5, 30, 10]
print(st.pop()," is popped") # [30, 10]
st.push("salam") # ["salam", 30, 10]
st.pop() # [30, 10]
print("Top of stack: ",st.top())
st.pop() # [10]
st.pop() # []
print("Stack is empty: ",st.is_empty())
```

صف عادی محدودیت هایی برای مدلسازی دارد. برای مثال فرض کنید که داریم صف نانوایی را مدل میکنیم اما نانوا آشناهایش را در ابتدای صف قرار می دهد!
برای همین لازم است که گاهی به سر صف هم عنصر اضافه شود یا حتی گاهی لازم است که از ته صف یک عنصر حذف شود(برای مثال یک نفر که حوصله ی صبر کردن ندارد چند نفر از آخر صف را منصرف می کند و سپس خودش یواشکی! به آخر صف اضافه می شود).
در واقع صف دو طرفه قابلیت های پشته و صف را به صورت همزمان پشتیبانی می کند.

<p></p>
یک صف دو طرفه باید از توابع زیر پشتیبانی کند:
<ul style="margin-top:0">
<li>$push\_front(x)$: عنصر $x$ را در ابتدای صف درج می کند.</li>
<li>$push\_back(x)$: عنصر $x$ را در انتهای صف درج می کند.</li>
<li>$pop\_front(x)$: عنصر ابتدای صف را حذف می کند و ان را باز می گرداند.</li>
<li>$pop\_back(x)$: عنصر انتهای صف را حذف و آن را باز می گرداند.</li>
<li>$front()$: عنصر ابتدای صف را باز می گرداند.</li>
<li>$back()$: عنصر انتهای صف را باز می گرداند.</li>
<li>$size()$: تعداد عناصر موجود در صف  را باز می گرداند.</li>
<li>$is\_empty()$: خالی بودن یا نبودن صف را مشخص می کند.</li>
<li>$is\_full()$: پر بودن یا نبودن صف را مشخص می کند.</li>
</ul>
تمامی اعمال در $O(1)$ انجام می‌شوند
<p></p>
پیاده سازی صف دوطرفه نیز به کمک آرایه یا لیست پیوندی انجام میشود.
<p></p>

<div style="direction:rtl;line-height:300%;"><font face="XB Zar" size=5>
<font size=6>
<p></p>
پیاده سازی صف دو طرفه به کمک آرایه
</font>
<p></p>
پیاده سازی صف دو طرفه با آرایه همانند پیاده سازی صف عادی با استفاده از آرایه است.
تنها تفاوت ها این است که هنگام اضافه شدن عنصر در ابتدای صف باید مقدار متغیر 
$first$
را یکی کاهش دهیم و 
$num$
را یکی افزایش دهیم.
در هنگام حذف عنصر از انتهای صف باید تنها مقدار
$num$
را یکی کاهش دهیم.

```python
class DoubleEndedQueue:
    def __init__(self, max_size):
        self.max_size = max_size
        self.Q = [0] * max_size
        self.num = 0
        self.first = 0
    
    def push_back(self, item): # Like one-way queue
        if self.num >= self.max_size:
            raise Exception("Queue overflow")
        self.Q[(self.num + self.first) % self.max_size] = item
        self.num += 1

    def push_front(self, item): # New
        if self.num >= self.max_size:
            raise Exception("Queue overflow")
        self.first = (self.first - 1) % self.max_size
        self.Q[self.first] = item
        self.num += 1

    def pop_front(self): # Like one-way queue
        if self.num == 0:
            raise Exception("Queue empty")
        item = self.Q[self.first]
        self.first = (self.first + 1) % self.max_size
        self.num -= 1
        return item
    
    def front(self): # Like one-way queue
        if self.num == 0:
            raise Exception("Queue empty")
        return self.Q[self.first]
        
    def pop_back(self): # New
        if self.num == 0:
            raise Exception("Queue empty")
        self.num -= 1
        return self.Q[(self.num + self.first) % self.max_size]
    
    def back(self): # New
        if self.num == 0:
            raise Exception("Queue empty")
        return self.Q[(self.num + self.first - 1) % self.max_size]
    
    def is_empty(self):
        return self.num == 0
    
    def size(self):
        return self.num
    
    def is_full(self):
        return self.num >= self.max_size
```

```python
#Example

dq=DoubleEndedQueue(7) # (front of queue)[](back of queue)
dq.push_front(1) # [1]
dq.push_back(2) # [1, 2]
dq.push_front(4) # [4, 1, 2]
print("queue size is: ",dq.size())
dq.push_back(0) # [4, 1, 2, 0]
print("back of queue is: ",dq.back())
dq.pop_front() # [1, 2, 0]
print("front of queue is: ",dq.front())
dq.pop_back() # [1, 2]
dq.pop_back() # [1]
dq.pop_back() # []
print("queue size is: ",dq.size())
```

و عنصر بعد از عنصر آخر نداریم و دیگری اینکه اگر لیست پیوندی ما حلقوی باشد عنصر بعد از عنصر آخر عنصر اول است و عنصر قبل از عنصر اول عنصر آخر است.
<p></p>
لیست پیوندی از توابع زیر پشتیبانی می کند:
<ul style="margin-top:0">
<li>$:insert\_after(data,x)$عنصر $data$ را پس از عنصر $x$ درج می کند.</li> 
<li>$:delete(x)$عنصر $x$ را حذف و آن را بازمیگرداند.</li>
<li>$:find(val)$ اولین عنصر با $data$ مساوی با $val$ را باز می گرداند.</li>
<li>$:get(ind)$ عنصر $ind$ ام لیست را باز میگرداند.</li>
<li>$:size()$ تعداد عناصر موجود در  لیست را باز می گرداند.</li>
<li>$:is\_empty()$ خالی بودن یا نبودن لیست را مشخص می کند.</li>
</ul>

درج در یک مکان آرایه از 
$O(n)$
است زیرا عناصر بعد از مکانی که عنصر جدید میخواهد در آنجا درج شود باید به جلو تغییر مکان دهند و تعداد آن ها از 
$O(n)$
است. اما برای درج یک عنصر مثل
$x$
در یک مکان مشخص از لیست پیوندی مانند درج در جایگاه
$i$
ام، کافیست که عنصر بعدی 
$i-1$
امین عنصر در صف را برابر
$x$
و عنصر بعدی 
$x$
را برابر عنصر 
$i$
ام قدیم قرار دهیم.
روند حذف نیز مشابه است.
به عنصر i
ام در آرایه دسترسی مستقیم داریم اما درلیست پیوندی باید از عنصر اول 
i بار جلو برویم
که بنابراین 
این عمل از 
$O(n)$
است.

یک کلاس برای $node$ های لیستمان می گیریم و در آن برای هر $node$ اطلاعات لازم را در $data$ نگه می داریم. همچنین $node$ بعد و قبل از آن را به ترتیب در $next$ و $prev$ آن قرار می دهیم.
حال به کمک کلاس $list$ که در آن $head$ به عنوان $node$ قبل از اولین عنصر و بعد از آخرین عنصر قرار دارد لیست پیوندی دوسویه را پیاده سازی می کنیم.
<br>
در واقع 
$head$
یک
$node$
با مقدار 
$None$
است که وظیفه اش نگه داشتن ابتدا و انتهای صف پیوندی ماست.
</font></div>

```python
class Node:
    
    def __init__(self, data):
        self.data = data  
        self.next = None  
        self.prev = None  
    
class List:
    def __init__(self):        
        self.head = Node(None)
        self.head.next = self.head
        self.head.prev = self.head
        self.n = 0
        
    def get(self,ind):
        if ind >=  self.size() : 
            raise Exception('Out of list')
        x = self.head.next
        for i in range(ind) :
            x=x.next
        return x
    
    def insert_after(self, x, data):
        y = Node(data)
        self.n += 1
        y.prev = x
        y.next = x.next
        x.next = y
        y.next.prev = y
        return y
        
    def delete(self, x):
        self.n -= 1
        x.prev.next = x.next
        x.next.prev = x.prev
        return x
        
    def find(self, val):
        x = self.head.next
        for i in range(self.size()) :
            if x.data == val :
                return x
            x=x.next
        return None
    
    def size(self):
        return self.n
      
    def is_empty(self):
        return self.n==0
```

```python
#Example

list=List() # head
list.insert_after(list.head,"milad") # head <-> milad
list.insert_after(list.get(0),"Arya") # head <-> milad <-> Arya
list.insert_after(list.find("Arya").prev,"jabbar") # head <-> milad <-> jabbar <-> Arya 
list.delete(list.find("jabbar")) # head <-> milad <-> Arya
print("Current size of list is",list.size())
```
