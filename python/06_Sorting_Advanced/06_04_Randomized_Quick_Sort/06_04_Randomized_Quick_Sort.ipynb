    <h1>
پیش نیاز تحلیل: احتمال
       </h1> 
       </font>
    <p></p>
یک شهود: ادعا کردیم که مرتب سازی سریع «در حالت میانگین» خوب عمل می‌کند. اما این به چه معناست؟ اولا باید دقت داشت که «میانگین یک پدیده‌ی تصادفی» یک تعریف ریاضی دقیق دارد. در ادامه با بیان تعریف های ابتدایی از احتمال، این تعریف دقیق را بیان خواهیم کرد. اما چیزی که اهمیت دارد اینست که عددی که مطابق این تعریف محاسبه می‌شود چه معنایی برای ما دارد. این معنا از ویژگی‌هایی که «میانگین یک پدیده‌ی تصادفی» (با همان تعریف ریاضی) دارد قابل استخراج است. این ویژگی‌ها به صورت ریاضی به دست می‌آیند و معمولا نتایج ساده‌ای نیستند. قبل از بیان تعریف ریاضی میانگین ویژگی‌هایی که از این تعریف قابل استخراج است را بیان می‌کنیم تا شهودی پیدا کنیم از اینکه چیزی که می‌خواهیم محاسبه کنیم اصلا چه معنایی دارد. این ویژگی‌ها را دقیق بیان نمی‌کنیم و صرفا معنای آن ها را در جهان واقع بیان می‌کنیم: (هر جا «میانگین پدیده‌ی تصادفی» نوشته شد به معنای تعریف ریاضی آن است.)
<br>
به طور نادقیق می‌توان گفت که اگر «میانگین یک پدیده‌ی تصادفی» 
$x$
باشد آنگاه اگر آن آزمایش را به تعداد زیاد تکرار کنیم و اعدادی که از آزمایش به دست آمده را با هم جمع کنیم و بر تعدادشان تقسیم کنیم (میانگین معمولی آن‌ها را حساب کنیم) این عدد به احتمال خیلی بالایی به  $x$
خیلی نزدیک خواهد بود. 
همچنین می‌توان گفت در هر بار انجام آزمایش عددی که به دست می‌آید به احتمال کمی فاصله خیلی زیادی (زیاد، نسبت به مقدار
$x$
)از
$x$
خواهد داشت. گرچه احتمال اینکه فاصله خیلی کمی داشته باشد هم لزوما زیاد نیست. 
<br>
ابتدا چند تعریف ابتدایی در احتمال بیان می‌کنیم سپس تعریف «میانگین» یا «امید ریاضی» را بیان می‌کنیم.
<ul>
<li> <b> فضای نمونه</b>: تمامی خروجی‌های ممکن یک رویداد، که آن را با $\Omega$ نشان می دهند و دو خاصیت زیر را دارد:
    <ol>
    <li> احتمال وقوع هر خروجی بزرگتر مساوی صفر است
    $$\exists i \in \Omega : p(i) \ge 0$$ </li>
    <li> مجموع احتمال خروجی ها برابر یک است
    $$\sum_{i \in \Omega} p(i) = 1$$ </li>
    </ol>
   برای مثال یک فضای نمونه می‌تواند همه دنباله‌های ممکن از حالت‌هایی باشد که آرایه از ابتدا تا زمان مرتب شدن پیدا می‌کند. می‌توانیم برای یک مسئله خاص در جهان واقع فضاهای نمونه‌ی متمایزی در نظر بگیریم. 
<li> <b>پیشامد</b>: هر زیرمجموعه‌ای از فضای نمونه را یک پیشامد می‌نامند. احتمال پیشامد $S \subseteq \Omega$ برابر است با
$$\sum_{i \in S} p(i) $$ </li>
<li> <b>متغیر تصادفی</b>:

در یک پدیده‌ی تصادفی ممکن است کمیت‌های گوناگونی وجود داشته باشند که علاقه‌مند باشیم بدانیم که به چه احتمالی چه مقداری می‌گیرند .
این علاقه انگیزه‌ی تعریف زیر است:
<br>
$X$ یک تابع از فضای نمونه به اعداد حقیقی است یا به عبارت دیگر 
  $X : \Omega \to \mathbb{R}$ 
<br>
به عنوان مثال اگر فضای نمونه همه دنباله‌های ممکن از حالت‌هایی باشد که آرایه از ابتدا تا زمان مرتب شدن پیدا می‌کند، یک متغیر تصادفی می‌تواند اندازه بخش‌هایی که در نتیجه افراز دوم در مرتب‌سازی سریع به دست می‌آید باشد.
یا در حالت کلی 
$X_i$
می‌تواند متغیر تصادفی اندازه بخش بزرگتر در نتیجه افراز $i$م باشد.
یا تعداد مقایسه‌هایی که انجام می‌شود خودش یک متغیر تصادفی است که به نظر می‌رسد داشتن اطلاعاتی راجع به اینکه به چه احتمالاتی برابر با چه مقادیری است درباره زمان اجرای الگوریتم پیام می‌دهد.
<br>
تمرین:مثال‌هایی از متغیر تصادفی در همین فضای نمونه مطرح کنید.</li>

<li> <b>امید ریاضی</b>:  تعریف دقیق آن به صورت زیر است:
    $$E[X] = \sum_{i \in \Omega} p(i).X(i)$$
    
    دقت کنید این تعریف به نوعی میانیگن وزن دار است روی فضای نمونه است.
به عنوان مثال امید ریاضی طول بخش اول که در اولین گام مرتب سازی سریع بر روی یک آرایه n عضوی اجرا می‌شود برابر با $n/2$ است. 
(چرا؟)</li>
<li> <b>خطی بودن امید ریاضی</b>: ادعا می‌کند که اگر $X_1$ تا $X_n$ متغیر تصادفی‌هایی باشند که روی $\Omega$ تعریف می شوند، آنگاه خواهیم داشت:
$$E[ \sum_{i=1}^{n} X_i] = \sum_{i=1}^{n} E[X_i]  $$
تمرین: این ویژگی را با کمک همین تعریف‌ها ثابت کنید. (این تعریف ها ساده شده هستند و در حد نیاز این بخش بیان شده‌اند)
</li>
<li> <b> تابع مشخصه </b> : یک نوع خاص از متغییر تصادفی که به صورت زیر تعریف می‌شود:
$$X : \Omega \to {0,1}$$
به عنوان مثال مقایسه شدن یا نشدن دو عنصر در روند مرتب‌سازی سریع
</li>
</ul>
</font>

    <h1>
مقدمات: تعریف مسئله به صورت احتمالاتی
</h1></font>
<p></p>
    
<b> فضای نمونه $\Omega$:</b> 
در هر مرحله یکی از اعداد آرایه را به عنوان محور انتخاب می‌کنیم. همه‌ی دنباله‌های ممکن از انتخاب محورها را به عنوان فضای نمونه تعریف می‌کنیم.  
<br>
<b> متغییر تصادفی $C(\sigma)$:</b> برای هر $\sigma \in \Omega$ برابر است با تعداد مقایسه‌ها هنگامی که دنباله محورها $\sigma$ 
باشد
.

<br>
زمان اجرای الگوریتم در اردر تعداد مقایسه‌های انجام شده بین اعداد آرایه است. بنابراین کافی است تعداد مقایسه‌ها را محاسبه کنیم.

پس اگر بخواهیم زمان اجرای مرتب‌سازی سریع را محاسبه کنیم، می‌توانیم مقدار $C(\sigma)$ را در حالت میانگین حساب کنیم، به عبارت دیگر $E[C(\sigma)]$ را به دست بیاوریم.
<br>
 اما این کار به صورت مستقیم بسیار سخت و پیچیده است

    <h1>
    روش طلایی: شکستن مسئله
    </h1>
</font>
<p></p>
هنگامی که با یک مسئله بسیار سخت روبه‌رو می شویم، می‌توانیم آن را به قطعات کوچکتر تقسیم کنیم، آن ها را حل کنیم و سپس با استفاده از آن‌ها به حل مسئله نهایی برسیم. (از این روش در درس طراحی الگوریتم‌ها نیز استفاده خواهید کرد.)
<br>
در این مسئله قطعات کوچکتر ما محاسبه‌ی تابع مشخصه است که به صورت زیر تعریف می‌شود:
<br>
$X_{ij} (\sigma) $= تعداد دفعاتی که $z_i$ و $z_j$ با هم مقایسه می‌شوند در حالی که دنباله‌ی محورها $\sigma$ است.
<br>
($z_i$ ، $i$امین عدد در آرایه مرتب شده‌است)
<br>
مقدار $X_{ij} (\sigma) $ صفر یا یک است، زیرا دو عنصر فقط هنگامی با هم مقایسه می‌شوند که یکی از آن‌ها محور باشد، و پس از این مقایسه عنصر دیگر در بخش جداگانه‌ای از عنصر محور قرار می‌گیرد، بنابراین هیچگاه دوباره با آن مقایسه نخواهد شد.
<br>
طبق تعریفی که برای $C(\sigma)$ ارائه شد، می توان تعداد مقایسه‌ها را این گونه بیان کرد که چند عنصر داریم که $X_{ij} (\sigma) $ آن‌ها یک است، یا به عبارت دیگر:
    $$ C(\sigma) =  \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} X_{ij}(\sigma)$$
حال با توجه به <b>خطی بودن امید ریاضی</b> داریم:
 $$ E[C(\sigma)] =  E[\sum_{i=1}^{n-1} \sum_{j=i+1}^{n} X_{ij}(\sigma)] = \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} E[X_{ij}(\sigma)]$$
 که نشان می دهیم می توان  $E[X_{ij}(\sigma)]$ را به راحتی به دست آورد

    <h1>
محاسبات ریاضی
    </h1>
</font>
    <p></p>
احتمال مقایسه شدن 
$z_i$ و $z_j$
را 
$p_{ij}$
می‌نامیم. 

 طبق تعریف 
 $X_{ij}$، $p_{ij}$
 داریم
 :
 
$$ E[X_{ij}] = p_{ij}.1 + (1-p_{ij}).0$$
بنابراین کافی است $p_{ij}$  را به دست آوریم.
<br>
گفتیم دو عدد فقط زمانی با هم مقایسه می‌شوند که یکی از آن‌ها محور باشد و هر دو در یک دسته باشند.
همچنین
هنگامی یکی از $z_i$ یا $z_j$ 
به عنوان محور انتخاب می‌شود، که قبلا هیچ یک از عناصر بینشان به عنوان محور انتخاب نشده‌باشد، در واقع عدد 
$z_i$ 
یا 
$z_j$
اولین عدد بین 
$z_i, z_{i+1} ,\dots, z_{j}$
باشد که به عنوان محور انتخاب می‌شود.

همچنین می‌دانیم که احتمال انتخاب  شدن همه‌ی اعضا با همدیگر برابر است. بنابراین خواهیم داشت:

$$p_{ij} = \frac{2}{j-i+1}$$
<br>
بنابراین داریم:
 $$ E[C(\sigma)] = \sum_{i=1}^{n-1} \sum_{j=i+1}^{n} \frac{2}{j-i+1} = 2\sum_{i=1}^{n-1} \sum_{j=i+1}^{n} \frac{1}{j-i+1}$$
 با استفاده از تغییر متغییر $k=j-i$ داریم
  $$E[X_{ij}]= 2\sum_{i=1}^{n-1} \sum_{k=1}^{n-i} \frac{1}{k+1} \leq   2\sum_{i=1}^{n-1} \sum_{k=2}^{n} \frac{1}{k} \leq 2n\sum_{k=2}^{n} \frac{1}{k}$$
که $\sum_{k=1}^{n} \frac{1}{k}$ به سری هارمونیک مشهور است و با استفاده از تقریب انتگرال ثابت می‌شود:
$$\sum_{k=1}^{n} \frac{1}{k} \leq ln(n)+1$$
در نتیجه خواهیم داشت:
$$E[C(\sigma)] \leq 2n\sum_{k=2}^{n} \frac{1}{k}\leq 2.n.ln(n)$$
بنابراین می‌توان نتیجه گرفت که مرتب‌سازی سریع از $\theta(nlogn)$ است.
<br>
تمرین: امیدریاضی تعداد نابه‌جایی ها در یک آرایه را محاسبه کنید (با فرض اینکه آرایه از میان جایگشت‌های مختلف اعضا به صورت تصادفی انتخاب شده‌است) سپس برای مرتب‌سازی درجی در حالت میانگین نتیجه گیری کنید.

    <h1>
انتخاب محور به صورت رندم

</h1>
</font>
<p></p>
دقت کنید در این تحلیل زمانی فرض شده‌‌است که آرایه به صورت تصادفی از بین همه جایگشت‌های ممکن اعداد به ما داده شده‌است بنابراین فرض شده احتمال اینکه هر عددی از زیر آرایه‌ای که مرتب‌ می‌کنیم انتخاب شود با هر عدد دیگری در آن زیر آٰرایه برابر است.(از این فرض در کجای اثبات استفاده شده است؟) حالا اگر چنین فرضی برای یک آرایه خاص درست نباشد تحلیل زمانی معتبر نخواهد بود. در این حالت با یک تغییر جزیی مشکل را رفع می‌کنیم: به جای اینکه عنصر انتهای آرایه را به عنوان محور انتخاب کنیم از تکنیکی به نام <b> نمونه‌برداری تصادفی </b> استفاده می‌کنیم: هر بار یک عنصر رندم را از بازه  به عنوان محور انتخاب می‌کنیم. بررسی کنید که در این حالت هم اثبات برقرار است و مستقل از اینکه  آرایه اولیه به چه احتمالی چه جایگشتی از آٰرایه مرتب‌شده باشد است.
