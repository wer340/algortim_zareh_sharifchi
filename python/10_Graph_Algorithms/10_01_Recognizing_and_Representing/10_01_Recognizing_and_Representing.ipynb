<div id="intro"> مقدمه</div>
</font>
گراف ها داده ساختار هایی پایه ای و پرکاربرد در حل انواع مسائل هستند. مسائل ساده ای مانند یافتن کوتاه ترین مسیر در یک شبکه تا مسائل پیچیده ای مانند تعیین بیشترین مقدار آبی که می توان از مجموعه ای از لوله ها عبور داد که هر کدام حداکثر ظرفیت مشخصی دارند
    (maximum-flow minimum-cut)
    ، می توانند به کمک گراف ها مدل شوند.  
    استفاده از داده ساختار مناسب در پیاده سازی و حل مسائل گراف از نکات مهمی است که باید به آن توجه کرد.
    در ادامه  این بخش به معرفی گراف ها و نمایش و پیاده سازی آن می پردازیم. 
</font></div>

<div id="section1"> تعاریف در گراف</div>
</font>
همانطور که می دانید، گراف داده ساختاری است که از دو جزء زیر تشکیل شده است 
<ul>
  <li>گره ها: مجموعه متناهی از راس ها. هر گره می تواند اطلاعات گوناگونی را در خود ذخیره کند(برای مثال یک id، اسم و غیره)</li>
  <li>یال ها: مجموعه متناهی از جفت راس های به هم متصل. در صورتی که گراف جهت دار باشد ترتیب را نیز در نظر می گیریم. همچنین در بعضی مسائل به هر یال یک عدد یا ارزش نسبت می دهیم که اصطلاحا به آن وزن یال می گویند.</li>
</ul>
    شکل زیر یک گراف جهت دار وزن دار با 10 راس را نشان می دهد.

<div id="section2">ذخیره سازی و نمایش گراف</div>
</font>
برای نمایش گراف دو روش رایج وجود دارد: 
<ul>
  <li>ماتریس مجاورت</li>
  <li>لیست مجاورت</li>
</ul>
    همچنین روش های دیگری همانند استفاده از ماتریس وقوع نیز وجود دارد. به صورت کلی نمایش و پیاده سازی یک گراف بسته به کاربردی که برای آن طراحی می شود می تواند متفاوت باشد. هر چه پیاده سازی ما باعث ساده شدن عملیات روی گراف شود، پیاده سازی بهتری محسوب می شود.  
در ادامه هر کدام از روش های فوق را توضیح می دهیم.

<font color=#FF7500 size=6>
<div id="section3">نمایش با ماتریس مجاورت</div>
</font>
ماتریس مجاورت یک ماتریس دو بعدی از 0 و 1 ها است که به تعداد راس های گراف، سطر و ستون دارد. فرض کنید این ماتریس را  $adj[][]$
     بنامیم . در این صورت $adj[][] = 1$
  اگر و تنها اگر یالی از راس $i$ ام به راس $j$
    ام وجود داشته باشد. بنابراین در صورتی که گراف جهت دار نباشد، ماترس مجاورت همواره متقارن خواهد بود. همچنین با تغییر کوچکی در ماترس مجاورت می توان از آن برای نمایش گراف های وزن دار استفاده کرد. به این صورت که $adj[][] = w$ اگر و تنها اگر یالی با وزن $w$ از راس $i$ ام به راس $j$
    ام وجود داشته باشد.
<img src="src/images/rep1.jpg" height="500" width="600">
در شکل بالا نمایش یک گراف 5 راسی با ماتریس مجاورت آن را مشاهده می کنید.
<br>
مزایا و معایب این نوع ذخیره سازی به شرح زیر است:
<ul>
  <li> مزایا: پیاده سازی با این روش ساده و واضح تر است. حذف یک یال در مرتبه زمانی  $O(1)$
    انجام پذیر است. همچنین بررسی وجود یک یال بین دو راس نیز از همین مرتبه زمانی امکان پذیر است.</li>
  <li>معایب: ذخیره سازی گراف با این روش به حافظه ی $O(V^{2})$ نیاز دارد. همچنین اضافه کردن یک راس به گراف نیز در مرتبه ی زمانی $O(V^{2})$ امکان پذیر است.</li>
</ul>
</font></div>

<div id="section4">نمایش با لیست مجاورت</div>
</font>
برای پیاده سازی در این روش از آرایه ای از لیست های پیوندی استفاده می شود. اندازه ی این آرایه برابر با تعداد راس های گراف است. این آرایه را $array[]$ می نامیم.
در این صورت $array[i]$ لیست پیوندی از راس هایی که یال جهت داری از راس $i$ ام به آن ها وجود دارد را نشان می دهد.
این نمایش همچنین می تواند برای گراف های وزن دار به کار گرفته شود. به طوری که وزن یال ها در راس های لیست های پیوندی ذخیره شود.
<img src="src/images/rep2.jpg" height="600" width="700">
در شکل بالا نمایش یک گراف 5 راسی با لیست مجاورت آن را مشاهده می کنید.
<br>
مزایا و معایب این نوع ذخیره سازی به شرح زیر است:
<ul>
  <li> مزایا: پیاده سازی با این روش حافظه ای از مرتبه ی $O(|V| + |E|)$ نیاز دارد. حداکثر تعداد یال های گراف برابر $\binom{V}{2}$ است بنابراین در بدترین حالت حافظه ی مورد نظراز $O(V ^ {2})$ است. همچنین افزودن یک راس معادل با اضافه کردن یک سطر جدید به آرایه است که نسب به ماتریس مجاورت ساده تر است. </li>
  <li>معایب: مرتبه زمانی بررسی وجود یال بین دو راس پیچیدگی $O(V)$ را دارد زیرا باید در لیست پیوندی مربوطه در آرایه، به جستجو پرداخت. بنابراین از این لحاظ بهینه نخواهد بود. </li>
</ul>
</font></div>

<div id="section5">نمایش با ماتریس وقوع</div>
</font>
برای پیاده سازی در این روش از آرایه ای دو بعدی استفاده می شود که تعداد سطر های آن برابر با تعداد راس های گراف و تعداد ستون های آن برابر با تعداد یال های گراف است. در صورتی که آرایه غیر جهت دارد باشد عضو های این ماتریس صفر و 1 هستند. اما برای گراف های وزن دار یا جهت دار می توان مشابه قسمت های قبل نمایش معادلی را ارائه کرد.در حالت ساده شده، اگر یال ها را نیز  شماره گذاری کنیم، در سطر $i$
     ام و ستون $j$
     ام ماتریس 1 قرار می دهیم اگر و تنها اگر یک سر یال 
   شماره $j$ راس $i$ قرار داشته باشد.
<img src="src/images/rep3.jpg" height="600" width="700">
در شکل بالا نمایش یک گراف 5 راسی با ماتریس وقوع آن را مشاهده می کنید . همانطور که دیده می شود ، چون این گراف جهت دار است، یال های خروجی از رئوس با $+1$ و یال های ورودی با 
    $-1$ نشان داده شده اند.
