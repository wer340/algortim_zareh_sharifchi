<b>
بررسی و اثبات الگوریتم</b>
</font> 
    <br>
   طبق تعریف باید بررسی کنیم به ازای هر یال
    $(u, v) \in E$
    راس u
    قبل از راس 
    v
    در الگوریتم ما ظاهر شود. پس کافی است نشان دهیم که زمان خاتمه 
    u 
    بیشتر از 
    v
    است. (چرا که در الگوریتم ما در این حالت است که u قبل از v ظاهر می‌شود)
    <br>
    مبحث 
    $Edge\; Classification$
    را به خاطر بیاورید.در آنجا به این نکته پی بردیم که برای یال‌های 
    $Forward\;Edge$
    و
    $Tree\;Edge$
    و
    $Cross\;Edge$
    این خاصیت برقرار است و تنها برای یال 
    $Back\;Edge$
    این خاصیت برقرار نیست. حال آنکه می‌دانیم اصلا 
    $Back\;Edge$
    نمی‌تواند موجود باشد زیرا اگر موجود باشد با بدون دور بودن گراف در تناقض است.(چرا؟)
    پس درستی الگوریتم ثابت شد.
</font></div>


<div style="direction:rtl;line-height:300%;" id="topo-algo"><font face="XB Zar" size=5>
<font color=#FF7500> 
<b>
تحلیل پیچیدگی زمانی الگوریتم
    </b>
</font> 
    زمان لازم برای اجرای هر خط را بررسی می‌کنیم
        <ol>
        <li> برای اجرای DFS به $O(m+n)$ زمان نیاز داریم</li>
        <li>در زمان $O(1)$ می‌توانیم عناصر را به ابتدای لیست اضافه کنیم.</li>
    </ol>
    پس در نهایت پیچیدگی زمانی الگوریتم از مرتبه
    $O(m+n)$
    می‌باشد که خطی است.
</font></div>


```python
from collections import defaultdict
class Graph:
    def __init__(self, N):
        self.graph = defaultdict(list) #Why we use this instead of {}? Can we write add_Edge function simpler?
        self.Vertices_num = N
        self.Answer_list = []

    def add_Edge(self, u, v):
        if u in self.graph:
            self.graph[u].append(v)
        else:
            self.graph[u] = [v]

    def DFS(self, v, visited):
        visited[v] = True
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFS(i, visited)

        self.Answer_list = [v] + self.Answer_list

    def TopologicalSort(self):
        visited = [False for _ in range(self.Vertices_num)]
        for i in range(self.Vertices_num):
            if visited[i] == False:
                self.DFS(i, visited)

        return self.Answer_list


DAG = Graph(8)
DAG.add_Edge(0,3)
DAG.add_Edge(1,3)
DAG.add_Edge(1,4)
DAG.add_Edge(2,4)
DAG.add_Edge(2,7)
DAG.add_Edge(3,5)
DAG.add_Edge(3,6)
DAG.add_Edge(3,7)
DAG.add_Edge(4,6)
DAG.TopologicalSort()
```

۲. در این مرحله تمامی رأس‌ها در داخل پشته قرار دارند.<br> 
۳. سپس گراف معکوس $G$ یعنی $G^T$ را به دست می‌آوریم. این بار گراف معکوس را با توجه به اولویت رأس‌ها در پشته، پیمایش می‌کنیم. به این ترتیب که در ابتدا از راسی شروع می‌کنیم که در بالای پشته قرار دارد. فرض کنید راس v در بالای پشته است. جستجوی عمق اول را از آن راس شروع می‌کنیم. فرض کنید در راسی به نام w به پایان برسد. تمام راس‌هایی که در این بین ملاقات شده جزو یک مؤلفه‌های قویا همبند هستند. تمام رأس‌های ملاقات شده را از پشته خارج و همین الگوریتم را بر روی راسی که در بالای پشته قرار دارد، تکرار می‌کنیم، تا زمانی که پشته خالی شود. <br>

```python
class Graph: 
   
    def __init__(self,vertices): 
        self.V= vertices
        self.graph = defaultdict(list)
   
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
   
    def DFSUtil(self,v,visited):  
        visited[v]= True
        print (v, end=" ")
        for i in self.graph[v]: 
            if visited[i]==False: 
                self.DFSUtil(i,visited) 
  
  
    def fillOrder(self,v,visited, stack): 
        visited[v]= True
        for i in self.graph[v]: 
            if visited[i]==False: 
                self.fillOrder(i, visited, stack) 
        stack = stack.append(v) 
      
    def getTranspose(self): 
        g = Graph(self.V) 
        for i in self.graph: 
            for j in self.graph[i]: 
                g.addEdge(j,i) 
        return g 
  
   
    def printSCCs(self): 
        stack = [] 
        visited =[False]*(self.V) 
        for i in range(self.V): 
            if visited[i]==False: 
                self.fillOrder(i, visited, stack) 
        gr = self.getTranspose() 
        visited =[False]*(self.V) 
        while stack: 
            i = stack.pop() 
            if visited[i]==False: 
                gr.DFSUtil(i, visited) 
                print()
   
# Create a graph given in the above diagram 
g = Graph(5) 
g.addEdge(1, 0) 
g.addEdge(0, 2) 
g.addEdge(2, 1) 
g.addEdge(0, 3) 
g.addEdge(3, 4) 
  
   
print ("Following are strongly connected components " +
                           "in given graph") 
g.printSCCs() 
```
