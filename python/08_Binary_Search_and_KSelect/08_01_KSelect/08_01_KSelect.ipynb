<h1><font color=#FF7500 id="selection" size=6>
الگوریتم پیداکردن k امین مرتبه آماری
</font></h1><p></p>
مساله‌ای که به آن می‌پردازیم، یافتن k امین کوچک‌ترین عدد در یک لیست ( k امین مرتبه‌ی آماری) است .

<br>
اولین راهی که برای این کار به ذهن می‌رسد ،مرتب کردن آرایه و انتخاب عنصر 
k
ام آن است که پیچیدگی زمانی 
$O(n log n )$
خواهد داشت.

ما در این دفترچه سعی می‌کنیم الگوریتم هایی با زمان اجرای بهتر برای این مسئله پیدا کنیم.
<p></p>
</font></div>

<h1><font color=#FF7500 id="quick" size=6>
انتخاب سریع
</font></h1><p></p>
انتخاب سریع از رویکرد یکسانی با مرتب‌سازی سریع بهره می‌برد، عضوی را به عنوان عنصر محوری انتخاب کرده و داده‌ها را بر پایهٔ آن به دو قسمت تقسیم می‌کند (دو قسمت کمتر و بیشتر از عنصر محوری).سپس با توجه به طول دو قسمت و k بازگشت روی یکی از قسمت‌ها انجام می‌دهد.
<br>
<div style="text-align:center" id="shape1">
<img src="src/images/quick.gif">
    <font size=3>شکل ۱: انتخاب سریع </font>
</div>
همانند مرتب‌سازی سریع، انتخاب سریع نیز دارای عملکرد حالت متوسط خوبی است، اما به شدت وابسته به عنصرهای محوری انتخاب شده می‌باشد. اگر عنصرهای محوری انتخاب شده مناسب باشند، به این معنا که تعداد داده‌ها را در هر مرحله با نسبت معینی کاهش دهند، تعداد داده‌ها به صورت نمایی کاهش می‌یابد و در نتیجه کارکرد الگوریتم به صورت خطی می‌شود. اگر عنصرهای محوری نامناسب به صورت پیاپی انتخاب شوند، مانند حالتی که در هر مرحله تنها یک عضو از داده‌ها کاسته شود، به بدترین عملکرد الگوریتم با 
$O(n^2)$
منجر خواهد شد.در بخش بعد برای پیدا کردن محور مناسب الگوریتم
<a href="#median">میانه میانه‌ها</a>
را معرفی می‌کنیم.
<br>
<br>
<font color=#FF7500> 
<b>
تمرین ۱: 
</b>
</font>
ثابت کنید پیچیدگی زمانی الگوریتم بالا به طور متوسط خطی است.
<br>
راهنمایی :
$$T(n) \leq cn$$
$$T(n) \leq an + \frac{1}{n}\sum_1^n{T(max(i, n - i - 1))}$$
<br>
<font color=#FF7500> 
<b>
تمرین ۲: 
</b>
</font>
چرا به طور متوسط پیچیدگی زمانی این الگوریتم از $O(n)$ است درحالی که در مرتب‌سازی سریع $O(n log n)$ است ؟
<br>
<br>
در ادامه پیاده‌سازی از الگوریتم بالا ارائه می‌دهیم.
خوب هست که جست‌و‌جو کنید و شبه کد‌های دیگر از جمله کتاب CLRS را ببینید. به عنوان مثال <a href="#shape1">شکل ۱</a> پیاده‌سازی این الگوریتم به صورت درجا است. پیاده سازی ارائه شده زیر کمی متفاوت هست.
<p></p>
</font></div>

```python
import random

def random_select(a, k):
    assert len(a) > k

    v = a[random.randint(0, len(a) - 1)]

    # less than, equal, greater than x
    lt = [x for x in a if x < v]
    eq = [x for x in a if x == v]
    gt = [x for x in a if x > v]

    if len(lt) > k:
        return random_select(lt, k)
    elif len(eq) + len(lt) > k:
        return v
    else:
        return random_select(gt, k - len(eq) - len(lt))

a = [5, 6, 0, 1, 10, 12, -1, 8, 100]

print (random_select(a, 0))
print (random_select(a, 2))
print (random_select(a, 5))
print (random_select(a, 8))
```

<h1><font color=#FF7500 id="median" size=6>
میانه میانه‌ها
</font></h1><p></p>
این الگوریتم برای پیدا کردن میانه‌ی آرایه و یا به طور کلی‌تر 
$i$ 
امین عنصر یک آرایه می‌باشد.
<br>

الگوریتم به صورت زیر می‌باشد:

<ul>
<li>
آرایه را به $n/5$ دسته‌ی ۵ تایی تقسیم می‌کنیم.
</li>
<li>
هر دسته را مرتب کرده و میانه‌ی آن را پیدا می‌کنیم.
</li>
<li>
به صورت بازگشتی میانه‌ی میانه‌های $n / 5$ دسته‌ی یاد شده را انتخاب می‌کنیم.
</li>
<li>
فرض کنید 
$x$
میانه ای باشد که در مرحله قبل بدست آورده ایم.
حال می توانیم آرایه را براساس 
$x$
به دو دسته اعداد کوچکتر از 
$x$
و اعداد بزرگتر از 
$x$
تقسیم کنیم و با توجه به اندازه‌ی دسته‌ی اول و مقدار
$i$
عدد مورد نظر را در دسته مناسب به صورت بازگشتی به کمک الگوریتم گفته شده 
پیدا کنیم.
</li>
</ul>

فرض کنید $x$ میانه ای باشد که در مرحله ۳ بدست آورده ایم. حال تمام دسته هایی که میانه آن ها از $x$ کمتر است را در نظر بگیرید تعداد آنها $n / 10$ است.

در هر یک از این دسته ها حداقل دو عضو کوچکتر از میانه خود دسته به همراه میانه انتخاب کرده از آن دسته مقداری کمتر از $x$ دارند که تعداد آن ها $3 n / 10$ است.
به طور مشابه می توان ثابت کرد حداقل $3n / 10$ عنصر بزرگتر از $x$ در آرایه وجود دارد.
<br>
<br>

<div style="text-align:center">
<img src="src/images/median2.png" width="90%">
    <font size=3>شکل ۲:پیچیدگی زمانی الگوریتم میانه میانه‌ها </font>
</div>
با توجه به اینکه دسته اعداد کوچک تر از $x$ و دسته اعداد بزرگتر از $x$ حداکثر $7n/10$ عضو دارند مرتبه زمانی الگوریتم فوق در بدترین حالت از فرمول زیر بدست می آید
:
$$T(n)\leq T(n/5)+T(7n/10)+c.n$$
که $T(n/5)$ زمان لازم برای پیدا کردن میانه میانه هاست
و $T(7n/10)$ زمان لازم برای حل مساله در مرحله ی بعد بازگشتی است.

از فرمول بالا می توان با استقرا دریافت که 
$T(n)$ از مرتبه زمانی $\theta(n)$ است.
<br>
همچنین با استفاده از درخت زیر مشاهده می‌شود که با طی کردن هر مرحله جمع رئوس نسبت به مرحله‌ی قبل ۹/۱۰ شده است و در نهایت مرتبه‌ی زبانی الگوریتم $\theta(n)$ خواهد بود.
<div style="text-align:center">
<img src="src/images/select-tree.png" width="60%">
    <font size=3>شکل ۳:درخت پیچیدگی زمانی الگوریتم میانه میانه‌ها </font>
</div>
<br>
<br>
<font color=#FF7500> 
<b>
تمرین ۳: 
</b>
</font>
در حالت دسته‌های ۳ و ۷ تایی تابع $T(n)$ را تعیین کنید و پیچیدگی زمانی آن را تعیین کنید.
<br>
راهنمایی : 
$$T_3(n)=T_3(n/3)+T_3(4n/6)+c.n = \theta(nlogn)$$
$$T_7(n)=T_7(n/7)+T_7(10n/14)+c.n =\theta(n)$$
<br>
<br>
در ادامه پیاده‌سازی از این الگوریتم ارائه می‌دهیم. 
<p></p>
</font></div>

```python
def median_of_medians(A, i):

    #divide A into sublists of len 5
    sublists = [A[j:j+5] for j in range(0, len(A), 5)]
    medians = [sorted(sublist)[len(sublist)/2] for sublist in sublists]
    if len(medians) <= 5:
        pivot = sorted(medians)[len(medians)/2]
    else:
        #the pivot is the median of the medians
        pivot = median_of_medians(medians, len(medians)/2)

    #partitioning step
    low = [j for j in A if j < pivot]
    high = [j for j in A if j > pivot]

    k = len(low)
    if i < k:
        return median_of_medians(low,i)
    elif i > k:
        return median_of_medians(high,i-k-1)
    else: #pivot = k
        return pivot

A = [1,2,3,4,5,1000,8,9,99]
B = [1,2,3,4,5,6]
print (median_of_medians(A, 0)) 
print (median_of_medians(A,7))
print (median_of_medians(B,4))
```

<p></p>
برای اطمینان پیدا کردن از عملکرد خطی حتی در بدترین حالت ، می‌توان از الگوریتم بالا برای تعیین عنصر محوری استفاده کرد.البته سربار محاسباتی تعیین عنصر محوری در این روش زیاد است و در عمل از این الگوریتم استفاده نمی‌شود.به همین دلیل می‌توان از 
<a href="https://en.wikipedia.org/wiki/Introselect">انتخاب درونگرا</a>
 استفاده کرد که با ترکیب انتخاب سریع با میانهٔ میانه‌ها، هم در حالت میانگین و هم در بدترین حالت، عملکردی خطی را نتیجه می‌دهد. 
 <br>
 همچنین با همین دیدگاه می‌توان روش مرتب سازی جدیدی به نام 
 <a href="http://www.geeksforgeeks.org/know-your-sorting-algorithm-set-2-introsort-cs-sorting-weapon/">
 مرتب‌سازی درونگرا
 </a>
 را معرفی کرد که سریع‌ترین الگوریتم موجود است.
 <br>
 مرتب‌سازی درونگرا یک مرتب‌سازی ترکیبی است که از روش‌های مرتب‌سازی سریع ، مرتب‌سازی هرمی و مرتب‌سازی درجی استفاده می‌کند.
 <br>

  روش این الگوریتم به این صورت است که در ابتدا با مرتب‌سازی سریع شروع می‌کند و اگر عمق بازگشتی آن از یک حد مشخصی گذشت به سراغ مرتب سازی هرمی می‌رود تا از گرفتار شدن در بدترین حالت الگوریتم مرتب‌سازی سریع که پیچیدگی زمانی برابر با
  $O(n^2)$
  دارد جلوگیری کند. و همچنین هر گاه که تعداد اجزای مرتب‌سازی کم باشد از مرتب‌سازی درجی استفاده می‌کند:
  
  <ul>
<li>
اگر عمق اجرای الگوریتم (عمق فراخوانی‌های تابع بازگشتی) حد بیشینه برای عمق را رد کند سراغ مرتب سازی هرمی می‌رود. حد بیشینه‌را برابر با 
$2*log(N)$
تعریف می‌کنیم.
</li>
<li>
اگر اندازه‌ی قطعه‌ای که قرار است مرتب کنیم به حدی کوچک باشد باشد که استفاده از مرتب‌سازی درجی بهینه تر باشد از آن استفاده می‌کنیم و این حد را برابر با ۱۶ تعریف می‌کنیم.
</li>
<li>
اگر هیچ کدام از این دو مورد رخ ندهد همان مرتب‌سازی سریع را ادامه خواهیم داد.
</li>
</ul>
